```python
import sys, importlib
import numpy as np
import pandas as pd

module_name = 'infer'
module_path = '/home/sequenzia/dev/solen-rl-project/utils/infer.py'

spec = importlib.util.spec_from_file_location(module_name, module_path)
infer = importlib.util.module_from_spec(spec)
sys.modules[module_name] = infer
spec.loader.exec_module(infer)

AGENTS_DIR = "/home/sequenzia/dev/solen-rl-project/trained-agents"

DATA_DIR = "/home/sequenzia/dev/solen-rl-project/data"

N_ENVS = 5
N_STEPS = 100000

NO_RENDER = True

all_infer_data = {}
```

# Results

```python
for key, data in all_infer_data.items():

    print(f"{key}")
    print(data.describe())
    print("\n")

    data.to_parquet(f"{DATA_DIR}/{key}.parquet")

```

```python
all_scores_list = []

for key, data in all_infer_data.items():
    
    for score in data['scores'].values.tolist():

        key_split = key.split("_")
        algo = key_split[0]
        rom = key_split[1]

        all_scores_list.append([key, algo, rom, score])

all_scores = pd.DataFrame(all_scores_list, columns=["run_key","algo","rom","score"])

```

```python
from sklearn.preprocessing import normalize
import plotly.express as px
```

```python
all_scores['score_norm'] = normalize(all_scores['score'].values.reshape(-1,1), axis=0, norm='l2') * 1000

breakout_scores = all_scores[all_scores['rom'] == 'BREAKOUT']
pong_scores = all_scores[all_scores['rom'] == 'PONG']
spaceinvaders_scores = all_scores[all_scores['rom'] == 'SPACEINVADERS']
mspacman_scores = all_scores[all_scores['rom'] == 'MSPACMAN']

a2c_scores = all_scores[all_scores['algo'] == 'A2C']
ppo_scores = all_scores[all_scores['algo'] == 'PPO']
dqn_scores = all_scores[all_scores['algo'] == 'DQN']

breakout_scores_summary = breakout_scores[['run_key','score']].groupby('run_key').agg(['count','mean', 'std', 'min', 'max'])
pong_scores_summary = pong_scores[['run_key','score']].groupby('run_key').agg(['count', 'mean', 'std', 'min', 'max'])
spaceinvaders_scores_summary = spaceinvaders_scores[['run_key','score']].groupby('run_key').agg(['count','mean', 'std', 'min', 'max'])
mspacman_scores_summary = mspacman_scores[['run_key','score']].groupby('run_key').agg(['count','mean', 'std', 'min', 'max'])

a2c_scores_summary = a2c_scores[['score_norm']].agg(['count','mean', 'std', 'min', 'max'])
ppo_scores_summary = ppo_scores[['score_norm']].agg(['count','mean', 'std', 'min', 'max'])
dqn_scores_summary = dqn_scores[['score_norm']].agg(['count','mean', 'std', 'min', 'max'])
```

```python
breakdown_cols = ['Algorithm', 'Min', 'Max', 'Mean', 'STD']

ppo_breakdown_vals = ['PPO', ppo_scores['score_norm'].min(), ppo_scores['score_norm'].max(), ppo_scores['score_norm'].mean(), ppo_scores['score_norm'].std()]
dqn_breakdown_vals = ['DQN', dqn_scores['score_norm'].min(), dqn_scores['score_norm'].max(), dqn_scores['score_norm'].mean(), dqn_scores['score_norm'].std()]
a2c_breakdown_vals = ['A2C', a2c_scores['score_norm'].min(), a2c_scores['score_norm'].max(), a2c_scores['score_norm'].mean(), a2c_scores['score_norm'].std()]

algo_breakdown = pd.DataFrame([ppo_breakdown_vals, dqn_breakdown_vals, a2c_breakdown_vals], columns=breakdown_cols)

algo_breakdown
```

```python
fig = px.bar(algo_breakdown, x="Algorithm", y="Mean", color="Algorithm", color_discrete_sequence=px.colors.qualitative.Dark24, title="Normalized Scores by Algorithm: Mean")
fig.show()
```

```python
fig = px.bar(algo_breakdown, x="Algorithm", y="Max", color="Algorithm", color_discrete_sequence=px.colors.qualitative.Dark24, title="Normalized Scores by Algorithm: Max")
fig.show()
```

```python
fig = px.bar(algo_breakdown, x="Algorithm", y="STD", color="Algorithm", color_discrete_sequence=px.colors.qualitative.Dark24, title="Normalized Scores by Algorithm: Standard Deviation")
fig.show()
```

## Game Scores

```python
games_breakdown = pd.DataFrame()

games_breakdown['Breakout'] = breakout_scores[['algo','score']].groupby('algo').mean()
games_breakdown['Pong'] = pong_scores[['algo','score']].groupby('algo').mean()
games_breakdown['SpaceInvaders'] = spaceinvaders_scores[['algo','score']].groupby('algo').mean()
games_breakdown['MsPacman'] = mspacman_scores[['algo','score']].groupby('algo').mean()

games_breakdown['Algorithm'] = games_breakdown.index

breakdown_cols = ['Breakout', 'Pong', 'SpaceInvaders', 'MsPacman']

fig = px.bar(games_breakdown, x=breakdown_cols, y="Algorithm", color_discrete_sequence=px.colors.qualitative.Vivid, title="Mean Scores of Algorithms by Game")
fig.show()

games_breakdown
```
